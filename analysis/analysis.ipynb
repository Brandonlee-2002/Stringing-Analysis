# Badminton Racket Restringing Data Analysis
### Objective:
- Analyze racket-string combinations to view player preferences 
- Explore tension distributions
- Identify Outliers 

### Dataset Information:
- **NameID**: Unique ID for each player
- **Racket** & **String**: Equipment used. Ex: AX88D Pro & BG66UM
- **Tension**: String Tension in lbs



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/Users/brandonlee/PycharmProjects/Badminton_Racket_Analysis/data/RestringLog.csv')
```

```python
# Explore the data
df.head()

# Check for missing values and duplicates
df.isnull().sum()
df = df.drop_duplicates()

# Perform string analysis (e.g., extracting year from the string)
df['extracted'] = df['String'].str.extract(r'(\d{4})')

# Frequency of unique strings
string_counts = df['String'].value_counts()

# Plotting the counts of unique strings
plt.figure(figsize=(10, 6))
string_counts.plot(kind='bar')
plt.title('Frequency of Strings')
plt.xlabel('String')
plt.ylabel('Frequency')
plt.show()

# Save the cleaned data
df.to_csv('processed_string_log.csv', index=False)
```
